### Accessing  elements from list
* To access values or elements in  list, we can use indexing or slicing. 
* Indexing starts from 0, which means the first element has an index of 0, the second element has an index of 1, and so on.
* We can  use negative indexing, which starts from the last element (-1), the second last element (-2), and so on.
### Indexing in  List
* Indexing is a technique for accessing the elements of a Python List. There are various ways by which we can access an element of a list.
### Positive Index
* In the case of Positive Indexing, the first element of the list has the index number 0, and the last element of the list has the index number N-1, where N is the total number of elements in the list (size of the list).


```python
x= [50,70,30,20,90,10,50]
print(x,type(x),id(x),sep='\n')
print(x[0])
```

### Negative Index
* Index -1 represents the last element and -N represents the first element of the list, where N is the length of the list


```python
x= [50,70,30,20,90,10,50]
print(x,type(x),id(x),sep='\n')
print(x[-1])
```

### Slicing
* Syntax Lst[ Initial : End : IndexJump ]
* If Lst is a list, then the above expression returns the portion of the list from index Initial to index End, at a step size IndexJump
* It can be  both with positive indexes as well as negative indexes. 


```python
x= [50,70,30,20,90,10,50]
print(x,type(x),id(x),sep='\n')
print(x[1:5])
```

### Range 
* range()populates a numbers within a given range 
* syntax for range() is : range(start,end,step)
  * start is starting number 
  * end is till end , not including end 
  * step is step size 
* default values of start is zero and step is +1

```python
r= range(1,11)  # till end means upto 10
print(r,type(r),hex(id(r)),sep='\n')
```

```python
list(r) #type casting is converting range to list
```

```python
r=range(10,1,-1)
print(r,type(r),hex(id(r)),sep='\n')
```

```python
list(r)
```

```python
list(range(0,11,1)),list(range(11)) #start is '0' and step is 1
```

### Adding Elements to a List

### Method 3:

### 2.pop()
* pop() is an inbuilt function that removes and returns the last value from the List or the given index value.
* pop() acts on index numbers 
* Syntax: list_name.pop(index)
##### parameters:
* index (optional) – The value at index is popped out and removed. If the index is not given, then the last element is popped out and removed.
##### Return:
* Returns The last value or the given index value from the list.
* Exception: Raises IndexError When the index is out of range.


```python
x=[1,2,3,4]
y=x.pop(2)
print(y)
```

```python
y=[78,98,68]
z=y.pop(1)
print(z)
```
